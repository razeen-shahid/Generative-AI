```python
import gradio as gr
from transformers import pipeline

# Load Whisper model for speech-to-text
asr_pipeline = pipeline("automatic-speech-recognition", model="openai/whisper-large-v3")

# Load Summarization model
summarization_pipeline = pipeline("summarization", model="facebook/bart-large-cnn")

# Function to transcribe and summarize
def transcribe_and_summarize(audio):
    # Speech-to-text with timestamp handling
    transcription_result = asr_pipeline(audio, return_timestamps=True)  # ✅ Fix added
    text = transcription_result["text"]

    # Generate summary (only if text is long enough)
    if len(text.split()) > 20:
        summary_result = summarization_pipeline(text, max_length=50, min_length=20, do_sample=False)
        summary = summary_result[0]["summary_text"]
    else:
        summary = "Text is too short for summarization."

    return text, summary

# Gradio UI
app = gr.Interface(
    fn=transcribe_and_summarize,
    inputs=gr.Audio(type="filepath"),
    outputs=[gr.Textbox(label="Text"), gr.Textbox(label="Summary")],
    title="Whisper AI Speech-to-Text & Summarization",
    description="Upload an audio file to get full transcription and a concise summary.",
)

# Run the app with PWA support
app.launch(pwa=True)
```
